```python
#https://stackoverflow.com/questions/48003907/how-to-train-naive-bayes-classifier-for-n-gram-movie-reviews
import nltk.classify.util
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import movie_reviews
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import ngrams

def create_ngram_features(words, n=2):
    ngram_vocab = ngrams(words, n)
    my_dict = dict([(ng, True) for ng in ngram_vocab])
    return my_dict

for n in [1,2,3,4,5]:
    pos_data = []
    for fileid in movie_reviews.fileids('pos'):
        words = movie_reviews.words(fileid)
        pos_data.append((create_ngram_features(words, n), "positive"))    

    neg_data = []
    for fileid in movie_reviews.fileids('neg'):
        words = movie_reviews.words(fileid)
        neg_data.append((create_ngram_features(words, n), "negative")) 

    train_set = pos_data[:800] + neg_data[:800]
    test_set =  pos_data[800:] + neg_data[800:]

    classifier = NaiveBayesClassifier.train(train_set)

    accuracy = nltk.classify.util.accuracy(classifier, test_set)
    print(str(n)+'-gram accuracy:', accuracy)
```

```python
from woe_utils import woe_wrapper

class woe_scikit_gnb(woe_wrapper):
    """
        WOE Wrapper around scikit GNB classifier
    """
    def __init__(self, clf, task=None, classes=None, input_size=None):
        super().__init__(clf, task, classes, input_size)
        self.cond_type = 'nb'
    
    def log_prob(self, x, y):
        """ 
        """
        # Lazy approach: subtract prior from joint_log_likelihood = log P(X |Y )  + log P(Y)
        #return self.model._joint_log_likelihood(x)[:,y] - self.model.class_prior_[y]        
        # If want to compute manually, can do:
        n_ij = - 0.5 * np.sum(np.log(2. * np.pi * self.model.sigma_[y, :]))
        n_ij -= 0.5 * np.sum(((x - self.model.theta_[y, :]) ** 2) /
                         (self.model.sigma_[y,:]), 1)          
        return n_ij     


    def log_prob_partial(self, subset, x, y):
        # Using naive Bayes, we do P(X_s | X_s1,...X_s, Y) = P (X_s, Y)
        # for GNB, this is easy: just drop non-included vars
        # en.wikipedia.org/wiki/Multivariate_normal_distribution#Marginal_distributions        
        n_ij = - 0.5 * np.sum(np.log(2. * np.pi * self.model.sigma_[y, subset]))
        n_ij -= 0.5 * np.sum(((x[:,subset] - self.model.theta_[y, subset]) ** 2) /
                         (self.model.sigma_[y, subset]), 1)          
        return n_ij
    
    def generalized_conditional_ll(self, x, y, subset=None, verbose = False, **kwargs):
        ## TODO: either make a non-torch version of the equivalent method in wrapper, or transsfortm to torch and use that one
        if subset is None:
            return self.log_prob(x, y)
        else:
            return self.log_prob_partial(subset, x, y)
#         elif self.cond_type == 'nb':

#             print('jere')
#             return self.model._joint_log_likelihood(x)[:,y] - self.model.class_prior_[y]
        

woe_model = woe_scikit_gnb(classifier, classes = [0,1])

print(woe_model.log_prob(X_test,0)[:10])

np.allclose(woe_model.log_prob(X_test,0), woe_model.log_prob_partial(np.arange(30), X_test, 0))
```

```python
# Compute partial and total woe

def explain(x):
    partial_woes = []
    for attribute in attributes:
        partial_woes.append(woe_model.woe(x, 1, 0, subset=attribute))

    partial_woes = np.array(partial_woes)
    total_woe = woe_model.woe(x, 1, 0)[0]
    prior_odds = np.log(classifier.class_prior_[1] - classifier.class_prior_[0])
    
    sorted_partial_woes= [(i,x[0]) for x,i in sorted(zip(partial_woes,range(len(attributes))), reverse=True)]
    print('Example index: {}'.format(idx))
    print('Prediction: {}'.format('positive' if classifier.predict(x)[0] else 'negative'))
    print('Bayes odds explanation:\n')
    print('   {:8.2f}     =        {:8.2f}  +  {:<8.2f}'.format(prior_odds + total_woe, prior_odds, total_woe))
    print('post. log-odds  =  prior log-odds  +  total_woe')
    
    print('\nTotal WoE in favor of Positive: {:8.2f}'.format(total_woe))
    
    for i,woe in sorted_partial_woes:
        print('woe({:>20}) = {:8.2f}'.format(attribute_names[i], woe))
        
    print('     {:>20} = {:8.2f}'.format('sum', partial_woes.sum()))

explain(X_test[idx:idx+1])
```
